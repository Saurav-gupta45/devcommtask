```python
spend_cols = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
df_copy['TotalSpend'] =df_copy[spend_cols].sum(axis=1)
```

```python
num_cols = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
for col in num_cols:
    df_copy[col] = df_copy[col].fillna(df_copy[col].median())
```

```python
df_copy['totalspend'] = df_copy[['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']].sum(axis=1)
```

```python
df_copy['NoSpend'] = (df_copy['TotalSpend'] == 0).astype(int)
```

```python
df_copy['Cabin'] = df_copy['Cabin'].fillna(df['Cabin'].mode()[0])
```

```python
col = ['HomePlanet','CryoSleep','Destination','VIP']
for col_no in col:
    df_copy[col_no] = df_copy[col_no].fillna(df_copy[col_no].mode()[0])
```

```python
df_copy['cabin_no1'] = df_copy['Cabin'].str.split('/').str[0]
df_copy['cabin_no2'] = df_copy['Cabin'].str.split('/').str[1]
df_copy['cabin_no3'] = df_copy['Cabin'].str.split('/').str[2]
```

```python
df_copy['cabin_no2'] = df_copy['cabin_no2'].astype(int)
```

```python
df_copy['Id'] = df_copy['PassengerId'].str.split('_').str[0]
```

```python
df_copy.head()
```

```python
no_of_person = df_copy['Id'].value_counts()
df_copy['no_of_person'] = df_copy['Id'].map(no_of_person)
```

```python
df_copy.head()
```

```python
df_copy['IsAlone'] = df_copy['no_of_person'] == 1
```

```python
df_copy.drop(['Name', 'Cabin', 'PassengerId', 'Id'], axis=1, inplace=True)
```

```python
df_copy.head()
```

```python
df_copy['no_of_person'].unique()
```

```python
df_copy['CryoSleep'] = df_copy['CryoSleep'].map({True:1 , False:0})
```

```python
df_copy['VIP'] = df_copy['VIP'].map({True:1 , False:0})
df_copy['IsAlone'] = df_copy['IsAlone'].map({True:1 , False:0})
```

```python
df_copy['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 18, 35, 60, 80], labels=['Child', 'Teen', 'YoungAdult', 'Adult', 'Senior'])
```

```python
from sklearn.preprocessing import LabelEncoder

label_cols = ['HomePlanet', 'Destination', 'cabin_no1','cabin_no3','AgeGroup']
le = LabelEncoder()
for col in label_cols:
    df_copy[col] = le.fit_transform(df_copy[col])
```
